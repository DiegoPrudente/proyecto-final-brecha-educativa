```python
import pandas as pd
pd.set_option('display.max_columns', None)

# 1. Leemos el archivo saltando las 5 primeras filas de títulos combinados
df_edu = pd.read_csv('../datos/escolaridad_inegi_bruto.csv', encoding='latin1', skiprows=5)

# 2. Renombramos todas las columnas a mano para tener control total
df_edu.columns = [
    'Estado', 
    'Edu_2000_Total', 'Edu_2000_H', 'Edu_2000_M',
    'Edu_2010_Total', 'Edu_2010_H', 'Edu_2010_M',
    'Edu_2015_Total', 'Edu_2015_H', 'Edu_2015_M',
    'Edu_2020_Total', 'Edu_2020_H', 'Edu_2020_M'
]

# 3. Limpiamos los textos
df_edu['Estado'] = df_edu['Estado'].str.strip()

# 4. Filtramos la basura
df_edu = df_edu[df_edu['Estado'] != 'Estados Unidos Mexicanos']
df_edu = df_edu.dropna().reset_index(drop=True)

print("✅ Educación limpia. Filas:", len(df_edu))
```

```python
# 1. Leemos el archivo totalmente en crudo
df_pib = pd.read_csv('../datos/pib_inegi_bruto.csv', encoding='latin1', header=None)

# 2. Extraemos ÚNICAMENTE las columnas que necesitamos por su índice
df_pib = df_pib[[0, 8, 13, 18]]

# 3. Renombramos
df_pib.columns = ['Estado', 'PIB_2010', 'PIB_2015', 'PIB_2020']

# 4. Cortamos la basura de arriba y reiniciamos índice
df_pib = df_pib.iloc[6:].reset_index(drop=True)

# 5. Limpiamos los textos
df_pib['Estado'] = df_pib['Estado'].str.strip()

# 6. Filtramos para quedarnos SOLO con los 32 estados válidos
estados_validos = df_edu['Estado'].tolist()
df_pib = df_pib[df_pib['Estado'].isin(estados_validos)].reset_index(drop=True)

print("✅ PIB limpio. Filas:", len(df_pib))
```

```python
# Unimos ambas tablas
df_final = pd.merge(df_edu, df_pib, on='Estado', how='inner')

# Guardamos el archivo final
df_final.to_csv('../datos/datos_educacion_pib_limpio.csv', index=False)

print("🚀 ¡Fusión exitosa! Archivo guardado.")
```
